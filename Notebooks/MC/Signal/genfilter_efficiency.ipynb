```python
from Notebooks.MC.Samples.signalnumbers import genfiltereff
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
mXX = [100, 150, 200, 500, 800, 1000]
mA = [0.25, 1.2, 5]
```

### 4mu

```python
## 300cm
d=[]
for ma_ in mA:
    row = []
    for mxx_ in mXX:
        row.append( genfiltereff['4mu'].get((mxx_, ma_, 300), np.nan) )
    d.append(row)
df = pd.DataFrame(d, index=mA, columns=mXX).transpose()
df
```

```python
fig, ax = plt.subplots(figsize=(7,5))
df.plot(kind='bar', ax=ax, rot=0, table=True)
fig.tight_layout()
ax.set_title('[4mu, lxy~300cm] Gen filter efficiency', x=0.0, ha="left")
ax.set_xlabel('mXX [GeV]', x=1.0, ha="left")
ax.set_ylabel('Efficiency', y=1.0, ha="right")
ax.get_legend().set_title('mA [GeV]')
ax.set_ylim([0,1])
ax.get_xaxis().set_visible(False)
ax.text(1.02,-0.03,'mXX [GeV]', transform=ax.transAxes)
ax.grid(True, axis='y', ls='--');
```
